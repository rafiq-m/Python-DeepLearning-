# Combining Data Overlapping

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
 .....: index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),
 .....: index=['f', 'e', 'd', 'c', 'b', 'a'])
```

```python
print(a)
b[-1] = np.nan
b
```

```python
np.where(pd.isnull(a), b, a)
```

# Next  example of Data Wrangling

```python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
 .....: 'b': [np.nan, 2., np.nan, 6.],
 .....: 'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
 .....: 'b': [np.nan, 3., 4., 6., 8.]})
print(df1)
df2
```

```python
df1.combine_first(df2) #jahan df1 mil raha wo lekar ae warna df2 ki value ko lekar ae
```

# Reshaping with Hierarchical Indexing

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
 .....: index=pd.Index(['Ohio', 'Colorado'], name='state'),
 .....: columns=pd.Index(['one', 'two', 'three'],
 .....: name='number'))
data
```

```python
result = data.stack()
result
```

```python
result.unstack()
```

```python
s1 = pd.Series([0, 1, 2, 3], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([4, 5, 6], index=['c', 'd', 'e'])
data2 = pd.concat([s1, s2], keys=['one', 'two'])
data2
```

```python
data2.unstack() #Column wise kardega or series ko data frame me convert kardega
```

```python
data2.unstack().stack(dropna = False)
```

```python
data2.unstack().stack()
```

# Pivot "Wide" To "Long"

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
 .....: 'A': [1, 2, 3],
 .....: 'B': [4, 5, 6],
 .....: 'C': [7, 8, 9]})
df
```

```python
melted = pd.melt(df, ['key'])
```

```python
melted
```

```python
reshaped = melted.pivot('key', 'variable', 'value')
reshaped     #Wapis se pichle data frame me convert hojaega
```
