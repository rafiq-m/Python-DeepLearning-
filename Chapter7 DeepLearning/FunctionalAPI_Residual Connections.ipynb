```python
from keras.datasets import boston_housing
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()
```

```python
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
test_data -= mean
test_data /= std
```

```python
from keras import models, Model
from keras import layers
import keras
def build_model():    
    i = layers.Input(shape=(train_data.shape[1],),dtype='float32')
    x = layers.Dense(64, activation='relu')(i)
    y = layers.Dense(128, activation='relu')(x)
    y = layers.Dense(128, activation='relu')(y)
    y = layers.Dense(64, activation='relu')(y)
    z = layers.add([y,x])
    z = layers.Dense(1)(z)
    model = Model(i,z)
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model
```

```python
model = build_model()
model.fit(train_data,train_targets,epochs=100,validation_data=(test_data, test_targets))
```

```python
model = build_model()
callbacks_list = [
        keras.callbacks.EarlyStopping(
        monitor='mae',
        patience=5,
        ),
        keras.callbacks.ModelCheckpoint(
        filepath='my_model.h5',
        monitor='mae',
        save_best_only=True,
        )]
model.fit(train_data,train_targets,epochs=100,callbacks=callbacks_list,
        validation_data=(test_data, test_targets))
```

```python
import keras
import numpy as np
class ActivationLogger(keras.callbacks.Callback):
            
    def on_epoch_end(self, epoch, logs=None):
        print("<<<<<< Epoch end >>>>>>>")
        
        
    def on_epoch_begin(self, epoch, logs=None):
        print("<<<<<< Epoch begin >>>>>>>")
        
        
    def on_train_begin(self, epoch, logs=None):
        print("<<<<<<< Training begin >>>>")

        
    def on_train_end(self, epoch, logs=None):
        print("<<<<<<< Training ended >>>>")
        
```

```python
callbacks_list = [ActivationLogger()]
model.fit(train_data,train_targets,epochs=10,callbacks=callbacks_list,
        validation_data=(test_data, test_targets))
```

```python
callbacks = [
keras.callbacks.TensorBoard(
log_dir='my_log_dir',
histogram_freq=1,
)]
history = model.fit(train_data, train_targets,
epochs=2000,
batch_size=128,
validation_split=0.2,
callbacks=callbacks)
```
